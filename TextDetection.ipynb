```python
import cv2
from pytesseract import pytesseract

pytesseract.tesseract_cmd = '/usr/bin/tesseract'
```

```python
img = cv2.imread('1.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
pytesseract.image_to_string(img)
bounding_boxes = pytesseract.image_to_boxes(img)
height, width = img.shape[0:2]
for bb in bounding_boxes.splitlines():
    bounding_box = bb.split(' ')
    x, y, w, h = [int(n) for n in bounding_box[1:5]]
    print([bounding_box[0], x, y, w, h])
    cv2.rectangle(img, (x, height - y), (w, height - h), (0,0,255), 1)
    cv2.putText(img, bounding_box[0], (x, height - y + 20), cv2.FONT_HERSHEY_COMPLEX, 1, (50,50,255),1)

cv2.imshow('result',img)
while cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()
    break
```

```python
height, width = img.shape[0:2]
img_cpy = cv2.imread('1.png')
boxes = pytesseract.image_to_data(img_cpy)
# boxes = boxes.splitlines()
# attributes = line_wise_data[0].split()
# records = [box.split() for box in line_wise_data[1:]]
for x, bounding_box in enumerate(boxes.splitlines()):
    if x!=0:
        bounding_box = bounding_box.split()
        if bounding_box[-1] != "-1": 
            x, y, w, h = [int(n) for n in bounding_box[6:10]]
            cv2.rectangle(img_cpy, (x, y), (x+w, y+h), (0,0,255), 1)
            cv2.putText(img_cpy, bounding_box[-1], (x, y+h+10), cv2.FONT_HERSHEY_COMPLEX, 0.3, (50,50,255),1)
            print(bounding_box)


cv2.imshow('result-1',img_cpy)
while cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()
    break
```
